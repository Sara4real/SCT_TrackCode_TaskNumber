```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset (skip first 4 rows of metadata)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv", skiprows=4)

# Check structure
df.head()
```

```python
# Step 3: Clean the '2022' column (convert to numeric)
df['2022'] = pd.to_numeric(df['2022'], errors='coerce')

# Drop rows with missing values in 2022
df_2022 = df.dropna(subset=['2022'])
```

```python
# Step 4: Sort and get Top 10 countries by population in 2022
df_top10 = df_2022.sort_values(by='2022', ascending=False).head(10)

# Preview
print(df_top10[['Country Name', '2022']])
```

```python
# Step 5: Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', data=df_top10, palette='viridis')
plt.title("Top 10 Countries by Population in 2022")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Step 6: Histogram for all countries' populations in 2022
plt.figure(figsize=(10,6))
plt.hist(df_2022['2022'], bins=30, color='purple', edgecolor='black')
plt.title("Distribution of Population in 2022")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
